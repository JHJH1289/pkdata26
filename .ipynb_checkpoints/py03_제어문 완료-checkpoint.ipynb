```python
for i in range(1, 10):
    print("==========")
    print(f"{i} 단")
    print("----------")
    for j in range(1, 10):
        num = i * j
        print(f"{i} X {j} = {num}")
    print("==========")
```

```python
for i in range(2, 10):
    num = []
    print("==========")
    print(f"{i} 단")
    print("----------")
    for j in range(1, 10):
        val = i * j
        num.append(val)
        print(val)
    print("==========")
```

```python
names = ['홍길동', '김철수', '이영희','이순신']
scores = [85, 92, 78, 99]
grade = ["합격" if x > 85 else "불합격" for x in scores]

for x,y,z in zip(names, scores,grade):
    print(f"{x} 님은 {y}점 이므로 {z}입니다.")
```
