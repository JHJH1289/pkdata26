```python
#sort - 리스트정렬
a = [5,43,4,5,56,3,2,4,1]
b = ['gf','f','u','a','e','f','z','c']
a.sort()
b.sort()
print(a)
print(b)
```

```python
#리스트 뒤집기
a = [5,43,4,5,56,3,2,4,1]
b = ['gf','f','u','a','e','f','z','c']
a.reverse()
b.reverse()
print(a,b, sep='\n')
print(a[::-1],b[::-1], sep='\n')
```

```python
#리스트 뒤집기
a = [5,43,4,5,56,3,2,4,1]
a.sort()
a.reverse()
print(a)
b = [5,43,4,5,56,3,2,4,1]
b.sort(reverse = True) # True는 내림차순, False 오름차순이 기본값
print(b)
```

```python
#sorted 응용하기
fruit =['banana','kiwi','orange','melon','apple','strawberry']
print(sorted(fruit))
print(sorted(fruit,key = len))
print(sorted(fruit,key = len, reverse=True))
```

```python
#- 선택 출력
# "바구니에 ['호박', '감자', '소주', '맥주', '피자', '치킨']이 있습니다."
#"꺼내고 싶은 물품을 입력하세요."
#"바구니에서 ??? 을 꺼내어 ['~', '~', '~', '~', '~']가 남았습니다"
    
basket = ['호박', '감자', '소주', '맥주', '피자', '치킨']
print(f"바구니에 {basket}이 있습니다.")
inp = input("꺼내고 싶은 물품을 입력하세요: ")
basket.remove(inp)
print(f"바구니에서 {inp}을(를) 꺼내어 {basket}가 남았습니다.")
```

```python
# 문제2
# 읽기전용 목록
# 입력받기1: "추가할 물건 이름을 적어주세요"
# 입력받기2: "어느 물품 뒤에 넣어둘까요?"
basket = ['호박', '감자', '소주', '맥주', '피자', '치킨']
print(f"바구니에 {basket}이 있습니다.")
inp_1 = input("추가할 물품을 입력하세요: ")
inp_2 = input("어느 물품 뒤에 넣어둘까요? :")
position = basket.index(inp_2)
basket.insert(position + 1, inp_1)  

print(basket)
```

```python
# access_logs = ["192.168.0.1", "192.168.0.2", "192.168.0.1", "10.0.0.1", "192.168.0.2", "192.168.0.5"]

'''
요구 사항:

access_logs를 set을 사용하여 중복을 제거하십시오.

새로운 관리자 IP "192.168.0.99"를 추가하십시오.

최종 고유 IP 개수를 출력하십시오.
'''
access_logs = ["192.168.0.1", "192.168.0.2", "192.168.0.1", "10.0.0.1", "192.168.0.2", "192.168.0.5"]
set_access_logs = set(access_logs)
set_access_logs.add("192.168.0.99")
print(len(set_access_logs))
```
