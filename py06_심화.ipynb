```python
class Cal:
    def __init__(self):
        self.result = 0

    def add(self, num):
        self.result += num
        return self.result
        
    def mul(self, num):
        self.result *= num
        return self.result
```

```python
class DDuck:
    def __init__(self, won, ea):
        self.won = won
        self.ea = ea

    def dduck(self):
        result = self.won * self.ea
        print(f"배가 고파서 떡볶이 집에서 떡볶이를 {result} 먹었다.")

    def odeng(self):
        result = self.won * 2 * self.ea
        print(f"배가 고파서 떡볶이 집에서 떡볶이를 {result} 원어치 먹었다.")
```

```python
재형 = DDuck(2500, 4)
호성 = DDuck(1500, 7)
재훈 = DDuck(2000, 5)
```

```python
재형.dduck()
재형.odeng()
```

```python
class Pet:
    def __init__(self, name, species):
        """펫을 생성"""
        self.name = name
        self.species = species
        self.hunger = 50

    def feed(self, many=10):
        """밥을 주면 배고픔 감소"""
        if self.hunger > 0:
            self.hunger -= many
            if self.hunger < 0:
                self.hunger = 0
            return f"{self.name}에게 밥을 줬어요. 배고픔 : {self.hunger}"
        return f"{self.name}는 이미 배가 불러요. 배고픔 : {self.hunger}"

    def play(self, how=20):
        """놀아주면 배고픔 증가"""
        if self.hunger < 100:
            self.hunger += how
            if self.hunger > 100:
                self.hunger = 100
        return f"{self.name}와 잘 놀았어요~!! 배고픔 : {self.hunger}"
```

```python
class Bacteria:
    def __init__(self, init_population, growth_rate):
        """세균초기화"""
        self.population = init_population
        self.growth = growth_rate
        self.time = 0 # 시작시간

    def grow(self, hours):
        """일정 시간이 지나면 세균이 증식"""
        self.time = hours
        self.population *= self.growth**hours
        return f"{self.time}시간이 흐릅니다. "

    def check(self):
        """현재 세균 수 확인"""
        print (f"현재 세균수 : {self.population}마리")
```

```python
# 대체제 클래스화
class Bank:
    interest_rate = 0.02  # 클래스 변수 p210

    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance or 0

    def bprint(self):  # 잔액조회
        return f"{self.owner} 님의 잔고는 {self.balance}원 입니다."

    def deposit(self, amount):  # 입금
        self.balance += amount
        return f"{self.owner} 님의 계좌에 {amount}원이 입금 되었습니다. 현재잔액 : {self.balance}원"

    def withdraw(self, amount):  # 출금
        if self.balance >= amount:
            self.balance -= amount
            return f"{self.owner} 님의 계좌에서 {amount}원이 출금 되었습니다. 현재잔액 : {self.balance}원"
        return "잔액이 부족합니다."

    def interest(self):
        total = self.balance * (1 + interest_rate)
        self.balance = total
        return f"{interest_rate}% 의 이자가 반영된 현재잔액 : {self.balance}원"
```

```python
acc1 = Bank("철수", 5000)
acc2 = Bank("영희", 1000)
acc3 = Bank("유진스", 50000)
acc4 = Bank("병우스님", 15000)
print(acc1.bprint())
print(acc2.bprint())
print(acc3.bprint())
print(acc4.bprint())
```

```python
# 예제5 응용
import random

class Character:
    def __init__(self, name, hp, attack_p):
        """캐릭터 초기화"""
        self.name = name
        self.hp = hp
        self.attack_p = attack_p

    def attack(self, opponent):
        """상대방을 공격"""
        damage = random.radint(2, self.attack_p)
        opponent.hp -= damage
        print(f"{self.name}가 {opponent.name}을 공격하여 {damage}를 입혔습니다.")

    def now_status(self):
        """현재 체력 상태 확인"""
        return f"이름:{self.name} 체력:{self.hp}"

    def is_alive(self):
        """캐릭터 생존여부 확인"""
        return self.hp > 0
```

```python
# 히어로와 몬스터 생성
player = Character("영웅", attack_p = 15, hp = 50)
monster = Character("골렘", 60, 9)
```

```python
# 전투 시작
import time as tt
print(f"{'전투시작':^30}")
while player.is_alive() and monster.is_alive():
    # input("경기를 시작하려면 엔터를 치세요.")
    tt.sleep(1)
    if player.is_alive(): player.attack(monster)
    if monster.is_alive(): monster.attack(player)
    print(player.now_status(), 'vs', monster.now_status())

# 결과
if player.is_alive():
    print("당신이 승리하셨습니다!!!")
else:
    print("패배하였습니다.. to continue..?")
```
