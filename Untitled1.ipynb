```python
#클래스 상속받고 메서드 추가하기 1

class UpgradeCalculator:
    def __init__(self):
        self.value = 0

    def add(self, val):
        self.value += val

    def minus(self, val):
        self.value -= val

cal = UpgradeCalculator()
cal.add(10)
cal.minus(7)

print(cal.value)
```

```python
#클래스 상속받고 메서드 추가하기 2

class MaxLimitCalculator:
    def __init__(self):
        self.value = 0

    def add(self, val):
        self.value += val
        if self.value >= 100: self.value = 100

    def minus(self, val):
        self.value -= val


cal = MaxLimitCalculator()
cal.add(50)
cal.add(60)

print(cal.value)
```

```python
# 음수 제거하기
# filter 와 lambda를 사용하여 리스트 [1,-2,3,-5,8,-3]에서 음수를 모두 제거해 보자.

nums = [1, -2, 3, -5, 8, -3]
result = list(filter(lambda x: x > 0, nums))
print(result)
```

```python
# abcd 문자열을 나열하는 경우의 수를 다음과 같이 모두 출력하시오 아 문제 안보고 노가다;

data = ['a', 'b', 'c', 'd']
length = len(data)

for i in range(length):
    for j in range(length):
        for k in range(length):
            for l in range(length):
                if i != j and i != k and i != l and j != k and j != l and k != l:
                    print(data[i] + data[j] + data[k] + data[l])
```

```python
# 다음 5명이 있고 할일은 3개라서 남는 2명은 휴식을 지정할 수 있는 프로그램작성
# 5명을 무작위로 섞어 앞의 3명은 일 나머지는 휴식
import random

people = ['김승현','김진호','강춘자','이예준','김현주']
todo = ['청소','빨래','설거지']

random.shuffle(people)

workers = people[:3]
rests = people[3:]

result = []

for i in range(len(todo)):
    result.append([workers[i], todo[i]])
    
for r in rests:
    result.append([r, '휴식'])

print(result)
```

```python
# 가로의 길이는 200cm 이고 세로의 길이는 80cm인 벽이 있다. 이 벽에 되도록 큰 정사각형 모양의 타일을 붙이려고 한다.
# 이때 붙이려는 타일 한 선의 길이와 붙이는 데 잎요한 타일의 개수를 구하시오.

import math

result = math.gcd(200, 80)
print(result)
```
