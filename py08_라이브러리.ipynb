```python
import os
import glob
import zipfile

f = 512 if not os.path.isdir("temp") else 0
if (f == 512):
    os.system("mkdir temp")
path = "./temp/"
i = 0
while i <= (ord("g") - 97):
    with open(path + "test_" + chr(97 + i)*2 + ".txt", "w") as f:
        f.write("test 중")
    i += 1

flist = glob.glob(path + '*')

with zipfile.ZipFile(path + 'all.zip','w') as z :
    for f in flist:
        z.write(f)
```

```python
# 멀티코어
import time
import multiprocessing

start = time.time()

def long_task():  # 5초의 시간이 걸리는 함수
    data = 0
    for i in range(100*1000*1000):
        # time.sleep(1)  # 1초간 대기한다.
        # print("working:%s\n" % i)
        data += i

print("Start")
processes = [ multiprocessing.Process(target=long_task) for i in range(20) ]

for p in processes:
    p.start()
for p in processes:
    p.join()

print("End")
print("processes 연산시간:", time.time() - start)
```
