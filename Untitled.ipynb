```python
# test
students = [("jane", 22, "A"), ("dave", 32, "B"), ("sally", 17, "B")]
title = ["name", "age", "grade"]

students_dic = []
for s in students:
    d = {}
    d[title[0]] = s[0]
    d[title[1]] = s[1]
    d[title[2]] = s[2]
    students_dic.append(d)
print(students_dic)

```
